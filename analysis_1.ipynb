```python
canal_zone = ip.load_json("canal_zone.json")
```

Create an Assembly

```python
canal_zone = ip.Assembly("canal_zone")
```

Since I have already demultiplexed, I just used  CLI to subset my data into a new folder with 74 samples; 69 LB & 5 BR
I am using relatively conservative and standard paramters for this assembly.
The only paramter I am really altering is a more conservative phred_Qscore ("set_params('phred_Qscore_offset', '43')") which is equivelant to saying no qscore lower than 30.

```python
canal_zone.set_params('project_dir', '/fs/project/adams.1970/cardenas.61/2019_analysis_final/Sept_analysis') 
canal_zone.set_params('sorted_fastq_path', './fastqs/*R1_.fastq.*')
canal_zone.set_params('assembly_method', 'denovo+reference')
canal_zone.set_params('datatype', 'ddrad')
canal_zone.set_params('reference_sequence', './Tzet_genomic.fna')
canal_zone.set_params('restriction_overhang', 'TGCAG, CGG')
canal_zone.set_params('max_low_qual_bases', '5')
canal_zone.set_params('mindepth_statistical', '6')
canal_zone.set_params('mindepth_majrule', '6')
canal_zone.set_params('clust_threshold', '0.85')
canal_zone.set_params('filter_adapters', '1')
canal_zone.set_params('filter_min_trim_len', '50')
canal_zone.set_params('max_Hs_consens','8,8')
canal_zone.set_params('min_samples_locus', '4')
canal_zone.set_params('max_SNPs_locus', '20, 30')
canal_zone.set_params('trim_reads', '0, 0, 0, 0')
canal_zone.set_params('output_formats', '*')
canal_zone.set_params('phred_Qscore_offset', '33')
canal_zone.get_params()
```

now we need to load the reads into this assembly

```python
canal_zone.run("1",ipyclient=yoink, force = True)
canal_zone.stats
```
